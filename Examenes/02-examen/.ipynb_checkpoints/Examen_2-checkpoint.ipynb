## 7) Crear un Dataframe con 10 registros y 6 columnas y Ordenar el dataframe 1 vez por cada columna, ascendente y descendente

```python
dt_10_6 = pd.DataFrame(np.random.randint(1,10,60).reshape(10, 6), columns=list('ABCDFE'))
dt_ascendente = dt_10_6.apply(lambda x: x.sort_values().values)
dt_descendente = dt_10_6.apply(lambda x: x.sort_values(ascending = False).values)
print("Ascendente")
print(dt_ascendente)
print("Descendente")
print(dt_descendente)
```

## 12) Crear un Dataframe con 10 registros y 6 columnas llenas de strings. Luego, unir la columna 1 y 2 en una sola, la 3 y 4, y la 5 y 6 concatenando su texto.

```python
df_strings = pd.DataFrame(pd.util.testing.rands_array(3, 60).reshape(10,6))
print(df_strings)

df_strings_final = pd.DataFrame(df_strings[0] + df_strings[1])
df_strings_final[1] = df_strings[2] + df_strings[3]
df_strings_final[2] = df_strings[4] + df_strings[5]
print(df_strings_final)
```

## 13) Crear un Dataframe con 10 registros y 6 columnas llenas de números randomicos del 1 al 10 enteros, obtener la frecuencia de repeticion de los numeros enteros en cada columna

```python
dt_10_6 = pd.DataFrame(np.random.randint(1,10,60).reshape(10, 6), columns=list('ABCDEF'))
for col in dt_10_6.columns:
    print("Columna" + str(col))
    print(dt_10_6[col].value_counts())
```

## 14) Crear un Dataframe con 10 registros y 3 columnas, A B C, llenas de números randomicos del 1 al 10 enteros. Crear una nueva columna con el calculo por fila (A * B ) / C

```python
dt_10_3 = pd.DataFrame(np.random.randint(1,10,30).reshape(10, 3), columns=list('ABC'))
#print(dt_10_3['A'][0])
datos = []
for index in dt_10_3.index:
    datos.append((dt_10_3['A'][index] * dt_10_3['B'][index]) / dt_10_3['C'][index])
#print(datos)
dt_10_3['Resultados'] = datos

print(dt_10_3)
```
